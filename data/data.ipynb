## Level of Measurements (Ölçülme Ölçeği)

4 farklı Level of Measurements vardır. Bunlar:
    
    Nominal
    
    Ordinal
    
    Interval
    
    Ratio
    
🟢Nominal:

    Kategori vardır.

    Sıralanamazlar.

    Mesela Asya ülkeleri: Çin Hindistan gibi. Bunlar birer kategoridir, sayılar yoktur. Dolayısıyla sıralanamazlar.

    Yada kadın/erkek gibi
    
🟢Ordinal:

    Sıralanabilir ama aralarındaki ilişki sayısal olarak anlamsızdır. 

    Mesela bir anket düşünün şıklar "asla, bazen, genelde, her zaman" olsun. Genelde bazenden daha genel bir ifade ama sayısal olarak ne kadar dediğimiz zaman bu sorunun cevabı yok. Bu tarz verilere ordinal veri denir.
    
🔵Nominal ve ordinal arasında sıralanabilirlik açısından fark var.

🟢Interval:

    Belirli bir ölçeği vardır.

    Natural zero starting point yoktur. Mesela sıcaklık için sıfırdan başlıyor yada 10 dereceden başlıyor diyemeyiz.

    Mesela oda sıcaklığı 20 derece 10 dereceden sıcaktır gibi karşılaştırmalarda bulunabiliriz.

🟢Ratio:

    True zero point vardır. Mesela yaş kilo gibi. Bir insanın yaşı sıfırdan başlar şeklinde zero point'i vardır.

    Belirli bir ölçeği vardır.
    
    Sıralanabilirler.

    Bazı bilim insanları interval ve ratio tiplerini tek bir tip olarak kabul eder.

## Central Tendency (Merkezi Eğilim)

Mean: Bir popülasyonun ortalamasına denir.

Median: bir listede ortada ki sayıya denir.

Mode: bir listede en çok bulunan sayıya denir.

Mean vs Median:

Mean outlier'lardan etkilenebilir. Bu nedenle median önemlidir. Mesela bir şirkette zam yapılmak istenirse çalışanların maaş ortalamasına mı bakılmalıdır? yada maaş medyanına mı bakılmalıdır?

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
import warnings
warnings.filterwarnings("ignore")

import numpy as np
from scipy import stats
```

```python
yas = [1,2,3,5,6,7,7,10,12,13]
# Mean
mean_yas = np.mean(yas)
print(mean_yas)
# Median
median_yas = np.median(yas)
print(median_yas)
# Mode
mode_yas = stats.mode(yas)
print(mode_yas)
```

[1,2,3,5,6,7,7,10,12,13] burada ortadaki sayı bulunmak istediğinde 5 tane sağda ve 5 tane solda sayı olduğunda, 6 ve 7 sayıları ortada kalmaktadır. Dolayısıyla 6 ile 7 sayısı arasında 6.5 olduğundan median değeri 6.5 olur.

Adet olarak en çok 7 sayısı bulunduğu için mod değeri 7'dir.

## Mean VS Median

Bazen mean kullanmak yerine median kullanmak daha mantıklı olabilir.

Mesela bir şirket düşünülem bu şirkette 10 çalışan var. Bu şirketin patronu eğer ortalama maaş 5 değerinin altındaysa her çalışana zam yapacak.

```python
salary = [1,4,3,2,5,4,2,3,1,500]
print("Mean of salary: ",np.mean(salary))
```

Yukarıdaki maaş listesine bakıldığında 5'e eşit veya altında olan 9 kişi vardır. 10 çalışandan 9'unun maaşı 5'in altındadır. Bu durumda zam yapılması uygun olur. Fakat birinin maaşı 500 ve ortalamayı büyük oranda etkilediği için, ortalamaya göre zam yapılamamaktadır.

Ortalama maaş 52.5 'dir. Patron bu ortalama yüzünden çalışanlarına çok maaş verdiğini düşünecektir ve bu nedenle maaşlarına zam yapmayacaktır.

Ama bildiğiniz gibi bu adil bir durum değildir. Çünkü burada 500 değerinde outlier diye adlandırabileceğimiz bir maaş değeri var.

Median bu outlier değerleri engeller.

```python
print("Median of salary: ",np.median(salary))
```

Median değeri 3'tür. Ve bu değer maaşların gerçek değerini yansıtıyor. 5'den de küçük olduğu için patron çalışanların maaşlarına zam yapacaktır.

y = np.random.uniform(5,8,100) satırında bulunan uniform ifadesi:

Eğer kod çıktısında bulunan görseldeki dağılımı görmeseydik, bu iki data için aynı diyebilirdik. Fakat görüldüğü üzere bu iki data'nın birbirleriyle pek bir alakası yok çünkü dağılımları farklıdır. Dolayısıyla sadece mean ve median değerlerine bakarak istatistiksel sonuçlar çıkaramayız. Bu yüzden Dispersion yani dağılıma bakmalıyız. ### Dispersion (Dağılım)

Dispersion dağılım demektir. "spread out" kavramı ile de ifade edilebilir.

Range: Bir listede en büyük ve en küçük sayı arasındaki farktır.

Variance: 

    varyans-değişiklik. Listedeki tüm noktaların ortalamadan ne kadar uzaklıkta olduğunu belirtir.


<br>
<img src="d2.png"  style="width:500px;height:200px"/>

Variance formülündeki terimler:

    X: Listedeki her bir noktadır.
    
    μ: Mean yani ortalamadır.
    
    (X-μ)^2 : Her bir noktadan ortalama çıkarılır ve - çıkma ihtimaline karşılık sonucun karesi alınır.
    
    N: Listenin eleman sayısıdır.
    
    N'e bölüm işlemi aslında normalize etmek gibi bir şeydir.

❗Variance formülünde: Bazı formüllerde N yerine (N-1) olabilir. Bu sample variance(N-1) ve population variance(N) diye iki farklı kavram olmasından kaynaklanıyor. Yani (N-1) görürseniz bu sample variance olarak adlandırılır. Eğer N görürseniz bu da population variance olarak adlandırılır.

Standard deviation (std):

    Variance'ın kare köküdür.
    
    İlerde göreceğimiz dağılımlarda bir dağılımın içindeki değerlerin ortalamaya olan uzaklıklarının ne kadar olduğunu belirtmek için kullanacağımız bir kavram. Aslında çok kullanılan ve önemli bir kavram.
    
    Standard deviation'u gaussian dağılımda çok kullanacağız. Standard deviation'u bulmak için variance'yi buluruz.

```python
# range
yas = [1,2,3,5,6,7,7,10,12,13]
print("Range: ", (np.max(yas)-np.min(yas))) #range: 13-1=12

# variance
#variance metodu ile:
print("Variance: ", (np.var(yas)))
#variance formülü ile:
var = sum((yas - np.mean(yas))**2)/len(yas)
print("Variance with formula: ",var)

# std
print("Std: ", (np.std(yas)))
std = np.sqrt(sum((yas - np.mean(yas))**2)/len(yas))
print("Std with formula: ",std)
```

## Quartiles

Q1 (lower quartile)(%25) : En küçük sayı ile median arasında kalan median değeridir.

Q2 (upper quartile)(%75) : Median ile en sonda kalan sayı arasında kalan median değeridir.

IQR(inter quartile range) = Q3-Q1

Outlier değerleri bulmak için IQR hesaplanmalıdır.

Outliers: Q1 - 1.5IQR and Q3 + 1.5IQR sınırları dışarısında kalan değerlerdir.

```python
# libraries
import pandas as pd
import seaborn as sns
plt.style.use("ggplot")
import warnings
warnings.filterwarnings("ignore")

# read data as pandas data frame
data = pd.read_csv("data.csv")
data = data.drop(['Unnamed: 32','id'],axis = 1)
```

```python
data.head()
```

```python
data_bening = data[data["diagnosis"] == "B"]
data_malignant = data[data["diagnosis"] == "M"]
desc = data_bening.radius_mean.describe()
Q1 = desc[4] #desc'in 4.indeksinde Q1 değeri vardır.
Q3 = desc[6] #desc'in 6.indeksinde Q3 değeri vardır.
IQR = Q3-Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
print("Anything outside this range is an outlier: (", lower_bound ,",", upper_bound,")")
data_bening[data_bening.radius_mean < lower_bound].radius_mean
print("Outliers: ",data_bening[(data_bening.radius_mean < lower_bound) | (data_bening.radius_mean > upper_bound)].radius_mean.values)
```

data_bening değişkenine iyi huylu tümörleri (B), data_malignant değişkenine kötü huylu tümörleri (M) atadık.

```python
melted_data = pd.melt(data,id_vars = "diagnosis",value_vars = ['radius_mean'])
sns.boxplot(x = "variable", y = "value", hue="diagnosis",data= melted_data)
plt.show()
```

İyi huylu ve kötü huylu data'nın radius_mean'ının box plotu çizdirilmiştir. Çizdirme işlemi için seaborn kütüphanesi kullanılmaktadır.

Dikdörtgenler içerisindeki çizgi median değerini ifade etmektedir.

## Pearson Correlation Coefficient (Pearson Korelasyon Katsayısı)

Pearson Correlation Coefficient: 
    
    Covariance'yi variable'lerin standart deviation'una bölümüdür. Aslında bunun anlamı standartize etmektir. Yani değerlerin küçük değerler arasına sıkıştırılması olarak düşünülebilir.
    
    Pearson Correlation Coefficient +1 ve -1 değerleri arasında değişir.

    +1 = positive linear correlation

    -1 = negative linear correlation

     0 = linear correlation yok

## Effect size

2 farklı grup arasındaki benzerliği ele almaktadır. Bakşa bir deyişle, iki farklı grup arasındaki farklılığı vurgulamaktadır. Yani 2 tane grup var ve bu 2 grup birbirine benziyor mu, benzemiyor mu bunun kararını effect size kullanarak verebilmemiz mümkün.

Cohen Effect Size: Eğer 2 grup arasındaki efect size 0.2'ye eşit ise bu iki grup birbirine benziyor demektir. Eğer 0.5 ise benzerlik var fakat çok fazla değil demektir. Eğer 0.8 ise iki grup birbirine benzemiyor demektir.

Veri setimizde kanser hücrelerini iyi huylu ve kötü huylu olmak üzere iki gruba ayırıyoruz. İyi huylu kanser hücrelerinin yarıçapları büyüklüklerini bir grup olarak, kötü huylu kanser hücrelerinin yarıçaplarını bir grup olarak düşüneceğiz. Bu iki grup arasındaki benzerliği cohen effect size'a göre tespit edeceğiz.

<font color="LightSeaGreen"><br>
Effect Size formülü:

<br>
<img src="d5.png"  style="width:200px;height:100px"/>

Formüldeki terimler:

Formül üzerindeki 1 ve 2 ifadeleri grupları ifade etmektedir.

M: Ortalama

n: Grupta kaç tane eleman olduğu

s: standart deviation

```python
mean_diff = data_malignant.radius_mean.mean() - data_bening.radius_mean.mean()    # m1 - m2
var_bening = data_bening.radius_mean.var() #iyi huylu tümörlerin yarıçaplarının varyansı
var_malignant = data_malignant.radius_mean.var() #kötü huylu tümörlerin yarıçaplarının varyansı
var_pooled = (len(data_bening)*var_malignant +len(data_malignant)*var_bening ) / float(len(data_bening)+ len(data_malignant))
effect_size = mean_diff/np.sqrt(var_pooled)
print("Effect size: ",effect_size)
```

Effect size değeri 1.9 elde edilmiştir. Bu değer oldukça büyüktür. Dolayısıyla bu iki grubun birbirinden farklı oldukları söylenebilir.
