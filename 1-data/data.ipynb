### Dispersion (Da캼l캼m)

Dispersion da캼l캼m demektir. "spread out" kavram캼 ile de ifade edilebilir.

Range: Bir listede en b칲y칲k ve en k칲칞칲k say캼 aras캼ndaki farkt캼r.

Variance: 

    varyans-de를를klik. Listedeki t칲m noktalar캼n ortalamadan ne kadar uzakl캼kta oldu릇nu belirtir.


<br>
<img src="d2.png"  style="width:500px;height:200px"/>

Variance form칲l칲ndeki terimler:

    X: Listedeki her bir noktad캼r.
    
    풮: Mean yani ortalamad캼r.
    
    (X-풮)^2 : Her bir noktadan ortalama 칞캼kar캼l캼r ve - 칞캼kma ihtimaline kar캼l캼k sonucun karesi al캼n캼r.
    
    N: Listenin eleman say캼s캼d캼r.
    
    N'e b칬l칲m i륿emi asl캼nda normalize etmek gibi bir 른ydir.

仇듔ariance form칲l칲nde: Baz캼 form칲llerde N yerine (N-1) olabilir. Bu sample variance(N-1) ve population variance(N) diye iki farkl캼 kavram olmas캼ndan kaynaklan캼yor. Yani (N-1) g칬r칲rseniz bu sample variance olarak adland캼r캼l캼r. E른r N g칬r칲rseniz bu da population variance olarak adland캼r캼l캼r.

Standard deviation (std):

    Variance'캼n kare k칬k칲d칲r.
    
    캻lerde g칬rece를miz da캼l캼mlarda bir da캼l캼m캼n i칞indeki de른rlerin ortalamaya olan uzakl캼klar캼n캼n ne kadar oldu릇nu belirtmek i칞in kullanaca캼m캼z bir kavram. Asl캼nda 칞ok kullan캼lan ve 칬nemli bir kavram.
    
    Standard deviation'u gaussian da캼l캼mda 칞ok kullanaca캼z. Standard deviation'u bulmak i칞in variance'yi buluruz.

```python
# range
yas = [1,2,3,5,6,7,7,10,12,13]
print("Range: ", (np.max(yas)-np.min(yas))) #range: 13-1=12

# variance
#variance metodu ile:
print("Variance: ", (np.var(yas)))
#variance form칲l칲 ile:
var = sum((yas - np.mean(yas))**2)/len(yas)
print("Variance with formula: ",var)

# std
print("Std: ", (np.std(yas)))
std = np.sqrt(sum((yas - np.mean(yas))**2)/len(yas))
print("Std with formula: ",std)
```

## Quartiles

Q1 (lower quartile)(%25) : En k칲칞칲k say캼 ile median aras캼nda kalan median de른ridir.

Q2 (upper quartile)(%75) : Median ile en sonda kalan say캼 aras캼nda kalan median de른ridir.

IQR(inter quartile range) = Q3-Q1

Outlier de른rleri bulmak i칞in IQR hesaplanmal캼d캼r.

Outliers: Q1 - 1.5IQR and Q3 + 1.5IQR s캼n캼rlar캼 d캼르r캼s캼nda kalan de른rlerdir.

```python
# libraries
import pandas as pd
import seaborn as sns
plt.style.use("ggplot")
import warnings
warnings.filterwarnings("ignore")

# read data as pandas data frame
data = pd.read_csv("data.csv")
data = data.drop(['Unnamed: 32','id'],axis = 1)
```

```python
data.head()
```

```python
data_bening = data[data["diagnosis"] == "B"]
data_malignant = data[data["diagnosis"] == "M"]
desc = data_bening.radius_mean.describe()
Q1 = desc[4] #desc'in 4.indeksinde Q1 de른ri vard캼r.
Q3 = desc[6] #desc'in 6.indeksinde Q3 de른ri vard캼r.
IQR = Q3-Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
print("Anything outside this range is an outlier: (", lower_bound ,",", upper_bound,")")
data_bening[data_bening.radius_mean < lower_bound].radius_mean
print("Outliers: ",data_bening[(data_bening.radius_mean < lower_bound) | (data_bening.radius_mean > upper_bound)].radius_mean.values)
```

data_bening de를륾enine iyi huylu t칲m칬rleri (B), data_malignant de를륾enine k칬t칲 huylu t칲m칬rleri (M) atad캼k.

```python
melted_data = pd.melt(data,id_vars = "diagnosis",value_vars = ['radius_mean'])
sns.boxplot(x = "variable", y = "value", hue="diagnosis",data= melted_data)
plt.show()
```

캻yi huylu ve k칬t칲 huylu data'n캼n radius_mean'캼n캼n box plotu 칞izdirilmi릆ir. 칂izdirme i륿emi i칞in seaborn k칲t칲phanesi kullan캼lmaktad캼r.

## Pearson Correlation Coefficient (Pearson Korelasyon Katsay캼s캼)

Pearson Correlation Coefficient: 
    
    Covariance'yi variable'lerin standart deviation'una b칬l칲m칲d칲r. Asl캼nda bunun anlam캼 standartize etmektir. Yani de른rlerin k칲칞칲k de른rler aras캼na s캼k캼릆캼r캼lmas캼 olarak d칲칲n칲lebilir.
    
    Pearson Correlation Coefficient +1 ve -1 de른rleri aras캼nda de를를r.

    +1 = positive linear correlation

    -1 = negative linear correlation

     0 = linear correlation yok

<br>

r: Correlation Coefficient

```python
p1 = data.loc[:,["area_mean","radius_mean"]].corr(method= "pearson")
p2 = data.radius_mean.cov(data.area_mean)/(data.radius_mean.std()*data.area_mean.std())
print('Pearson correlation: ')
print(p1)
print('Pearson correlation: ',p2)
```

## Spearman Rank Coefficient (Spearman Rank Katsay캼s캼)

Spearman Rank Coefficient ise outlier'lar ile ba르 칞캼kabilen daha dayan캼kl캼 bir metottur.

```python
ranked_data = data.rank() 
spearman_corr = ranked_data.loc[:,["area_mean","radius_mean"]].corr(method= "pearson")
print("Spearman's correlation: ")
print(spearman_corr)
```

data, rank edilerek ranked_data de를륾enine atanm캼릆캼r.
